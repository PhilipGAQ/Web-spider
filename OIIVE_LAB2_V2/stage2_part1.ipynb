```python
# Assuming douban2fb.txt is a mapping file with movie ID to entity ID
mapping_file = 'douban2fb.txt'

# Create a set to store the entity IDs from douban2fb.txt for faster lookup
#entity_ids存储的是douban2fb.txt中的entity_id,即初始的500多个实体
entity_ids_list = []
ids_to_entity = {}
with open(mapping_file, 'r') as mapping:
    for line in mapping:
        # Assuming the file format is "movieID\tentityID"
        movie_id, entity_id = line.strip().split()
        entity_id="<http://rdf.freebase.com/ns/" + entity_id+ ">"
        entity_ids_list.append(entity_id)
        ids_to_entity[entity_id] = movie_id
print(len(entity_ids_list))
```

```python
print(ids_to_entity)
```

```python
import gzip
from tqdm import tqdm
entity_list = []
relation_list = []

with gzip.open('second_3.gz','rb') as f:
    for line in tqdm(f,desc="Processing lines", unit="line"):
        triplet = line.strip().decode().split('\t')[:3]
        entity_list.append(triplet[0])
        relation_list.append(triplet[1])
        entity_list.append(triplet[2])

entity_list = list(set(entity_list))
relation_list = list(set(relation_list))

print("Number of entities: {}".format(len(entity_list)))
print("Number of relations: {}".format(len(relation_list)))
```

```python
#测试程序，检测entity_list是否包含entity_ids_list中的所有实体
for entity in entity_ids_list:
    if entity not in entity_list:
        print(entity)
    
```

```python
entity_part2_to_id = {}#实体到id的映射，0-577,即entity_list中的实体到id的映射
movie_id_to_id = {}
with open ('movie_id_map.txt','r') as f:
    for line in f:
        movie_id,entity_id = line.strip().split()
        movie_id_to_id[movie_id] = entity_id
#print(movie_id_to_id)
for entity in entity_ids_list:
    movie_id = ids_to_entity[entity]
    id = movie_id_to_id[movie_id]
    entity_part2_to_id[entity] = int(id)

```

```python
#去除578个初始实体后的所有实体的映射，数值从578开始
#在entity_list中去除entity_ids_list中的实体
i = 578
for entity in entity_list:
    if entity not in entity_ids_list:
        entity_part2_to_id[entity] = i
        i += 1
#将entity_part2_to_id按照id的大小排序
#entity_part2_to_id = sorted(entity_part2_to_id.items(), key=lambda item:item[1])
print(entity_part2_to_id)
```

```python
#建立关系的映射
i = 0
relation_to_id = {}
for relation in relation_list:
    relation_to_id[relation] = i
    i += 1

print(relation_to_id)
```

```python
#将三元组转换为id三元组，利用entity_part2_to_id和relation_to_id
import gzip
from tqdm import tqdm
second_3_id = []
with gzip.open('second_3.gz','rb') as f:
    for line in tqdm(f,desc="Processing lines", unit="line"):
        triplet = line.strip().decode().split('\t')[:3]
        second_3_id.append([entity_part2_to_id[triplet[0]],relation_to_id[triplet[1]],entity_part2_to_id[triplet[2]]])
```

```python
print(second_3_id)
```

```python
#将id三元组写入文件kg_final.txt
with open('kg_final.txt','w') as f:
    for triplet in second_3_id:
        f.write(str(triplet[0])+'\t'+str(triplet[1])+'\t'+str(triplet[2])+'\n')
```
